```python
from pandas import *
import numpy as np
import daniel_utils as du
%load_ext autoreload
%autoreload 2
```

```python
df = read_csv('csv/london.csv')
df = du.process_df(df)
df.to_csv('csv/processed_london.csv')
no_calc = 0
diff_calc = 0
for h in range(0, len(df) - 1):
#for h in range(0, 1):
    calc_length_preov = 0
    for i in range(1, 91):
        six_days = []
        three_days = []
        for j in range(0, 6):
            six_days.append(df['TEMP'+str(i+j)][h])
        for k in range(6, 9):
            three_days.append(df['TEMP'+str(i+k)][h])
        if max(six_days) < min(three_days):
            calc_length_preov = i + 5
            break
    if(str(calc_length_preov) != str(int(df['L_PREOVULATION'][h]))):
        if(calc_length_preov == 0):
            no_calc += 1
        else:
            diff_calc += 1
        print("Index: " + str(h))
        print("CALCULATED: " + str(calc_length_preov))
        print("RECORDED: " + str(int(df['L_PREOVULATION'][h])))
        print("---")
print("Total number of cycles: " + str(len(df)))
print("Total diffs: " + str(no_calc+diff_calc))
print("No Pre-Ov calculated: " + str(no_calc))
print("Different calculated value: " + str(diff_calc))
```
